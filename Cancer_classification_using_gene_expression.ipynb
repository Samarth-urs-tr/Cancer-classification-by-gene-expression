```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder = LabelEncoder()
label_df['Cancer_encoded'] = label_encoder.fit_transform(label_df['cancer'])
```

```python
#Assigning the labels for the train and test data
```

```python
train_data_x = train_data.reset_index(drop=True)
train_data_y = label_df[label_df.patient <= 38].reset_index(drop=True)
```

```python
test_data_x = test_data.reset_index(drop=True)
test_data_y = label_df[label_df.patient > 38].reset_index(drop=True)
```

```python
train_data_y = train_data_y.iloc[:,2]
train_data_y = train_data_y.tolist()
```

```python
train_data_y
```

```python
test_data_y = test_data_y.iloc[:,2]
test_data_y = test_data_y.tolist()
```

```python
test_data_y
```

```python
#Applying scaling
train_data_x = train_data_x.astype(float, 64)
test_data_x = test_data_x.astype(float, 64)

scaler = StandardScaler()
train_data_x_scaled = scaler.fit_transform(train_data_x)
test_data_x_scaled = scaler.fit_transform(test_data_x) 
```

```python
#Applying PCA to reduce the number of features
from sklearn.decomposition import PCA
```

```python
pca = PCA()
train_X_scaled = train_data_x_scaled
pca.fit_transform(train_X_scaled)
```

```python
total_variance = sum(pca.explained_variance_)
no_features = 0
current_variance = 0

while current_variance / total_variance < 0.95:
    current_variance += pca.explained_variance_[no_features]
    no_features += 1

print(f'{no_features} features explain around 95% of the variance')
```

```python
def pca_data(no_components, train_data_x_scaled, test_data_x_scaled):
    pca = PCA(n_components = no_components)
    train_X_pca = pca.fit_transform(train_data_x_scaled)
    train_X_pca = pd.DataFrame(train_X_pca)
    
    test_X_pca = pca.transform(test_data_x_scaled)
    test_X_pca = pd.DataFrame(test_X_pca)
    data = (train_X_pca, test_X_pca)
    
    return data
```

```python
#Applying logistic regression, decision tree, support vector machine (SVM), and random forest classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
def log_reg(train_X, train_y, test_X):
    classifier = LogisticRegression()
    classifier.fit(train_X, train_y)
    
    predictions = classifier.predict(test_X)
    
    return predictions

def dec_tree(train_X, train_y, test_X):
    classifier = DecisionTreeClassifier(random_state=0)
    classifier.fit(train_X, train_y)
    
    predictions = classifier.predict(test_X)
    
    return predictions

def random_forest(train_X, train_y, test_X):
    classifier = RandomForestClassifier(random_state=0)
    classifier.fit(train_X, train_y)
    
    predictions = classifier.predict(test_X)
    
    return predictions

def svm(train_X, train_y, test_X):
    classifier = SVC()
    classifier.fit(train_X, train_y)
    
    predictions = classifier.predict(test_X)
    
    return predictions
```

```python
#Using 32 features
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

train_X_32, test_X_32 = pca_data(32, train_data_x_scaled, test_data_x_scaled)

LR_32 = log_reg(train_X_32, train_data_y, test_X_32)
acc_LR_32 = accuracy_score(test_data_y, LR_32)
print("Accuracy of Logistic regression model with 32 features: ",acc_LR_32)
print("\n")
LR_32_cm = confusion_matrix(test_data_y, LR_32)
print("Confusion matrix of Logistic regression model with 32 features:", LR_32_cm)
print("\n")

dec_tree_32 = dec_tree(train_X_32, train_data_y, test_X_32)
acc_dec_tree_32 = accuracy_score(test_data_y, dec_tree_32)
print("Accuracy of Decision tree model with 32 features: ",acc_dec_tree_32)
print("\n")
dec_tree_32_cm = confusion_matrix(test_data_y, dec_tree_32)
print("Confusion matrix of Decision tree model with 32 features:", LR_32_cm)
print("\n")

RF_32 = random_forest(train_X_32, train_data_y, test_X_32)
acc_RF_32 = accuracy_score(test_data_y, RF_32)
print("Accuracy of Random Forest model with 32 features: ",acc_RF_32)
print("\n")
RF_32_cm = confusion_matrix(test_data_y, RF_32)
print("Confusion matrix of Random Forest model with 32 features:", RF_32_cm)
print("\n")

SVM_32 = svm(train_X_32, train_data_y, test_X_32)
acc_SVM_32 = accuracy_score(test_data_y, SVM_32)
print("Accuracy of SVM model with 32 features: ",acc_SVM_32)
print("\n")
SVM_32_cm = confusion_matrix(test_data_y, SVM_32)
print("Confusion matrix of SVM model with 32 features:", SVM_32_cm)
print("\n")
```
